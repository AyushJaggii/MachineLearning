# Customer Churn Prediction 

### Importing the Data

```python
import numpy as np 
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

### Understanding the Data

```python
dataset = pd.read_csv(r'E:\1. Tech\Machine Learning Projects\Customer Churn Prediction\Churn_Modelling.csv')
```

```python
dataset.head()
```

```python
## Assigning X & Y values

X = dataset.iloc[:,3:-1]
y = dataset.iloc[:,-1].values
```

```python
X.head()
```

```python
y
```

```python
print(X['Geography'].value_counts())
print(X['Gender'].value_counts())
print(X['NumOfProducts'].value_counts())
print(X['HasCrCard'].value_counts())
print(X['IsActiveMember'].value_counts())
```

```python
## Value Counts for Numpy 
unique = np.unique(y)
value_count = np.bincount(y)
for unique, value_count in zip(unique, value_count):
    print(f"{unique}: {value_count}")
```

### Manipulating Data

```python
X = pd.get_dummies(X, drop_first = True)
## Here the Categorical values are reduces by a column in order to reduce the complexity using the drop_first Parameter.
```

```python
X.head()
```

```python
## Splitting into Training and testing:


""" What Stratify will do is it will make sure the model is not bias, as the value counts of y are 0: 7963
1: 2037, since there so many counts of a single value thus, the stratify function will take the 85 & 15 percent train_test data from
0 values and 85 & 15 train_test data from 1 values, So, that both labels have equal distribution.
"""
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.15, stratify=y ,random_state=0)
```

```python
X_train
```

```python
## Standard scaling:

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
X_train
```

### Creating and training the model

```python
## Estimators are number of tress in a randomforest.
model_rf = RandomForestClassifier(n_estimators=100)
model_rf.fit(X_train, y_train)
```

```python
y_pred = model_rf.predict(X_test)
```

### Perfromance Evaluation of the Model

```python
confusion_matrix = confusion_matrix(y_test, y_pred)
```

```python
confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```
